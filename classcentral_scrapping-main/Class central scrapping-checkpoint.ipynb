```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
from urllib.request import Request, urlopen


url = "https://www.classcentral.com/"
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
# page = urlopen(url)
webpage = urlopen(req).read().decode("utf-8")
page_soup = soup(webpage, "html.parser")

# html = page.read().decode("utf-8")
# soup = BeautifulSoup(html, "html.parser")
```

```python
links_containers = page_soup.findAll("a")
dict1={}
link_lst=[]
for i in links_containers:
    link = i['href']
    text = i.text.strip()
#     if link[0]=='/':
#         link = "https://www.classcentral.com"+link

#     if text =='RSS Feed':
#         pass
#     req = Request(link , headers={'User-Agent': 'Mozilla/5.0'})
#     webpage_links = urlopen(req).read().decode("utf-8")
#     page_soup_links = soup(webpage_links, "html.parser")
#     tag_links = page_soup_links.body
#     link_content=[]
#     for j in tag_links.strings:
#         s=j.strip()
#         if s=="":
#             pass
#         else:
#             link_content.append(s)
    dict1={'link text': text,'link url':link}
    link_lst.append(dict1)
    print(len(link_lst))

```

```python
tag = page_soup.body
lst=[] 
# Print each string recursively
for string in tag.strings:
    s=string.strip()
    lst.append(s)
```

```python
final_list = [i for i in lst if i]
final_list
```
